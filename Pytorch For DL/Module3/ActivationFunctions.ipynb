# Activation Function

A function that enables neural networks to learn nonlinear (complex) relationships between features and the label.
**In short is a nonlinear transform of a neuron's output value before it is passed as an input to the next layer**

```python
import torch
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles
import pandas as pd
```

```python
device = "cuda" if torch.cuda.is_available() else "cpu"
device
```

```python
A = torch.arange(-10, 10, 1, dtype=torch.float32)
A
```

```python
plt.plot(A);
```

In Neural Networks, activation fucntion manipulates the weighted sum of all the inputs to a single neuron

so it goes like

several neurons -> single neuron in another layer(calculate weighted sum) -> output -> activation function

Stacking nonlinearities on nonlinearities lets us model very complicated relationships between the inputs and the predicted outputs

# RELU

Rectified Linear unit

Converts negative values to 0 while keeping positive values unchanged.

```python
def relu(x):
  return torch.maximum(torch.tensor(0), x) 
```

```python
plt.plot(relu(A));
```

# SIGMOID
any math function graph has a characteristic S-shaped or sigmoid curve

Converts any input number into a value between 0 and 1.

Interpreted as probability in binary classification tasks.

```python
def sigmoid(x):
  return 1 / (1 + torch.exp(-x))
```

```python
plt.plot(sigmoid(A));
```

# SoftMax
Converts a list of numbers to probabilities.

The probabilities add up to 1

Interpreted as how likely each option is.

```python
def softmax(x):
    exp_logits=torch.exp(x-torch.max(x))
    return exp_logits/torch.sum(exp_logits)
```

```python
plt.plot(softmax(A));
```

```python
torch.sum(A)
```

```python
# check if it adds up to 1
torch.sum(softmax(A))
```

# Tanh
Converts input values into numbers between -1 and 1.

Interpreted as helping the model understand both positive and negative signals, making learning often faster and better.

```python
def tanh(x):
    exp_pos = torch.exp(x)
    exp_neg = torch.exp(-x)
    return (exp_pos - exp_neg) / (exp_pos + exp_neg)
```

```python
plt.plot(tanh(A))
```
